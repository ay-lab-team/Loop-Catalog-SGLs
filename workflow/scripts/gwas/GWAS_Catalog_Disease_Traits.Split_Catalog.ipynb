## Getting all the OLS Link IDs

```python
link_ids = []
for uris in gwas_meta['MAPPED_TRAIT_URI']:
    if type(uris) == str:
        for uri in uris.split(','):
            link_ids.append(uri.strip())
link_ids = sorted(set(link_ids))
num_files = np.ceil(len(link_ids) / 100)
```

```python
# split the IDs into different files
split_dir = os.path.join(outdir, 'splits/')
os.makedirs(split_dir, exist_ok=True)
curr_bin = 0 
curr_fhandle = open(os.path.join(split_dir, 'split.ols_id.{}.txt'.format(curr_bin)), 'w')

for i, link_id in enumerate(link_ids):    
    if np.floor(i / 100) != curr_bin:
        curr_fhandle.close()
        curr_bin += 1 
        curr_fhandle = open(os.path.join(split_dir, 'split.ols_id.{}.txt'.format(curr_bin)), 'w')
    curr_fhandle.write(link_id + '\n')
    
```

```python
# close the last handle
curr_fhandle.close()
```
