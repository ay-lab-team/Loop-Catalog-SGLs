```python
import os
import subprocess
import pandas as pd 
os.chdir('<project-dir>')
outdir = 'results/refs/ensembl/'
os.makedirs(outdir, exist_ok=True)

fn = os.path.join(outdir, 'gencode.v30.annotation.gtf.gz')
if not os.path.exists(fn):
    url = 'https://ftp.ebi.ac.uk/pub/databases/gencode/Gencode_human/release_30/gencode.v30.annotation.gtf.gz'
    cmd = 'wget -O {} {}'.format(fn, url)
    output = subprocess.check_output(cmd, shell=True)
```

```python
gtf = pd.read_table(fn, header=None, skiprows=5)
```

```python
# extract the main columns
parsed = gtf.iloc[:, [0,3,4,8,6,2]]
```

```python
parsed
```

```python
# get the gene id and gene name
parsed['gene_id'] = parsed.iloc[:, 3].str.extract('gene_id "(ENSG[0-9.]+)"')
parsed['gene_name'] = parsed.iloc[:, 3].str.extract('gene_name "([A-Za-z0-9-\._]+)"')
parsed['gene_type'] = parsed.iloc[:, 3].str.extract('gene_type "([A-Za-z0-9-\._]+)"')


# final data cleanup
parsed = parsed.iloc[:, [0,1,2,4,5,6,7,8]]
```

```python
# use the gene ID as the name when the name is not 
# available
new_genenames = []
for ID, name in parsed.values[:, 5:7]:
    if type(name) == str:
        new_genenames.append(name)
    else:
        new_genenames.append(ID)
parsed.loc[:, 'gene_name'] = new_genenames
```

```python
parsed
```

```python
# save the final output
bed = os.path.join(outdir, 'gencode.v30.annotation.w_genetypes.bed')
parsed.to_csv(bed, sep='\t', header=None, index=False)
```
