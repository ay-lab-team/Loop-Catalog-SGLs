### Make longrange files for SGLs

```python
# load sgl annotations 
fn = 'results/main/sgls/ldpairs/T1D_34012112_Gaulton/GRCh37/Quach_2016/monocyte_IAV/monocyte_naive/master.tsv'
fn = 'results/main/sgls/ldpairs/T1D_34012112_Gaulton/GRCh37/Quach_2016/monocyte_R848/monocyte_naive/master.tsv'
tpl = 'results/main/GRCh37/sgls/ldpairs/*/*/*/*/script_version/master.tsv'
for fn in glob.glob(tpl):
        
    sgl_annos = pd.read_table(fn)

    # extract SNP-gene pairs which form an SGL
    sgls = sgl_annos.loc[(sgl_annos.has_fithichip_loop == 1)]

    # add SNP bin location
    snp_bins = sgls.snp_pos.apply(coordinates.get_snp_bin, res=res).tolist()
    snp_bins = np.array(snp_bins)
    sgls.loc[:, 'snp_start_bin'] = snp_bins[:, 0]
    sgls.loc[:, 'snp_end_bin'] = snp_bins[:, 1]

    # addd TSS bin location
    tss_start_bin = sgls.tss_start.apply(coordinates.get_closest_bin_start_end, res=res)
    tss_end_bin = sgls.tss_end.apply(coordinates.get_closest_bin_start_end, res=res)
    sgls.loc[:, 'tss_start_bin'] = tss_start_bin
    sgls.loc[:, 'tss_end_bin'] = tss_start_bin + res

    sgls_washu = sgls[['chrom', 'snp_start_bin', 'snp_end_bin', 
                      'chrom', 'tss_start_bin', 'tss_end_bin']]
    sgls_washu.loc[:, 'dummy_pval'] = 0.00001

    basename = fn.split('/')[2:-1]
    basename = '.'.join(basename)
    basename = '{}.longrange.bed'.format(basename)
    washu_fn = os.path.join(tracks_dir, basename)
    washu.bedpe_to_WashU_longrange(sgls_washu, washu_fn)
```

### Make a hub for all these 

```python
hub_jsons = []
```

```python
curr_hub = os.path.join(hubs_dir, 'loops/')
os.makedirs(curr_hub, exist_ok=True)
```

#### Add Genes

```python
gene_refbed = 'results/refs/washu/gencode.washu.refbed.gz'
gene_refbed_basename = os.path.basename(gene_refbed)
gene_refbed_link = os.path.join(curr_hub, gene_refbed_basename)
if not os.path.exists(gene_refbed_link):
    os.link(gene_refbed, gene_refbed_link)
    os.link(gene_refbed + '.tbi', gene_refbed_link + '.tbi')

gene_refbed_json = {'type': 'refbed',
                  'filename': gene_refbed_basename, 
                  'name': 'Gencode V30 - GRCh37',
                  'options': {'height': 100}}
hub_jsons.append(gene_refbed_json)
```

#### Add LongRanges

```python
tpl = os.path.join(tracks_dir, '*.longrange.bed.gz')
for track_fn in glob.glob(tpl):   
    basename = os.path.basename(track_fn)
    name = '.'.join(basename.split('.')[2:7])
    json_str = {'type': 'longrange',
                'filename': os.path.basename(track_fn),
                'name': 'SGL Pairs for {}'.format(name),
                'options': {'displayMode': 'arc', 'color':'purple', 'height': 200}}
    
    # symlink the track over
    track_link = os.path.join(curr_hub, basename)
    if not os.path.exists(track_link):
        os.link(track_fn, track_link)
        os.link(track_fn + '.tbi', track_link + '.tbi')

        
    hub_jsons.append(json_str)
```

```python
ignore_loop_sources = ['TFH', 'TH1', 'TH2', 'TH17', 'THSTAR', 'TREGMEM', 'TREGNAIVE']

tpl = 'results/main/2021_Nikhil_eQTL/Data/FitHiChIP_Loops/*/FitHiChIP_S/FitHiChIP.interactions_FitHiC_Q0.01_WashU.bed.gz'
for track_fn in glob.glob(tpl): 
    basename = os.path.basename(track_fn)
    loop_source = track_fn.split('/')[5]
    new_basename = '{}.{}'.format(loop_source, basename)
    
    if loop_source in ignore_loop_sources:
        continue
    
    # symlink the track over
    track_link = os.path.join(curr_hub, new_basename)
    if not os.path.exists(track_link):
        os.link(track_fn, track_link)
        os.link(track_fn + '.tbi', track_link + '.tbi')    
    
    json_str = {'type': 'longrange',
                'filename': os.path.basename(track_link),
                'name': 'Loops from {}'.format(loop_source),
                'options': {'displayMode': 'arc', 'color':'red', 'height': 200}}
    

    hub_jsons.append(json_str)
```

```python
hub_json_fn = os.path.join(curr_hub, 'hub.config.json')
with open(hub_json_fn, 'w') as f:
    f.write(json.dumps(hub_jsons, indent=4))
```
